## Los datos

Se nos a dado acceso a la base de datos de reservas del hotel en un archivo llamado 'hotel_bookings.csv', cuyó contenido es el siguiente:

| Columna     | Descripción              |
|------------|--------------------------|
| `Booking_ID` | Identificador unico de las reservas. |
| `no_of_adults` | El número de adultos. |
| `no_of_children` | El número de niños. |
| `no_of_weekend_nights` | Número de noche en fin de semana (Saturday o Sunday). |
| `no_of_week_nights` | Número de noches en la semana (Monday a Friday). |
| `type_of_meal_plan` | Tipo de plan de comidas incluido en la reserva. |
| `required_car_parking_space` | Si se requiere lugar de estacionamiento. |
| `room_type_reserved` | El tipo de cuarto reservado. |
| `lead_time` | Número de días antes de la fecha de llegada que se hizo la reserva. |
| `arrival_year` | Año de llegada. |
| `arrival_month` | Mes de llegada. |
| `arrival_date` | Fecha del mes de llegada. |
| `market_segment_type` | Como fue hecha la reservación. |
| `repeated_guest` | Si el huesped se a quedado en el hotel con anterioridad. |
| `no_of_previous_cancellations` | Número de cancelaciones previas. |
| `no_of_previous_bookings_not_canceled` | Número de reservas previas que fueron canceladas. |
| `avg_price_per_room` | Precio promedio de reserva por día. |
| `no_of_special_requests` | Número de solicitudes especiales que fueron hechas como parte de las reserva. |
| `booking_status` | Si la reserva fue cancelada o no. |

Source (data has been modified): https://www.kaggle.com/datasets/ahsan81/hotel-reservations-classification-dataset
